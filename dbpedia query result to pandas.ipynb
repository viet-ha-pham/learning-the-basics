```python
%%sparql http://dbpedia.org/sparql/ -qs capitals_in_europe
PREFIX dct: <http://purl.org/dc/terms/>
PREFIX dbc: <http://dbpedia.org/resource/Category:>
PREFIX dbp: <http://dbpedia.org/property/>

SELECT DISTINCT ?name ?capital ?populationTotal WHERE {
    ?capital dct:subject dbc:Capitals_in_Europe ;
             dbp:populationTotal ?populationTotal ; 
             foaf:name ?name. 
}
ORDER BY DESC(?populationTotal)
LIMIT 5
```

```python
capitals_in_europe.plot.barh('name', 'populationTotal').invert_yaxis()
```

```python
%%sparql http://dbpedia.org/sparql/ -qs Japanese_mystery_writers
PREFIX dct: <http://purl.org/dc/terms/>
PREFIX dbr: <http://dbpedia.org/resource/>
PREFIX dbc: <http://dbpedia.org/resource/Category:>
PREFIX dbp: <http://dbpedia.org/property/>
PREFIX dbo: <https://dbpedia.org/ontology/>
PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#> 

SELECT DISTINCT ?writer_name ?book_name WHERE {
    ?writer dct:subject dbc:Japanese_mystery_writers;
            rdfs:label ?writer_name.
    ?book dbp:author ?writer;
        rdfs:label ?book_name.
    
    FILTER (LANG(?writer_name)='en' && LANG(?book_name)='en')
}
```
